```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carga desde un archivo .xlsx sin indice
df = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df
```

```python
#Verificamos informacion del DataFrame
df.info()
```

**FILTROS BÁSICOS PARA VARIABLES CUALITATIVAS**

**filtros básicos para variables cualitativas**

```python
#Filtro por Objeto
#Se filtró la columna NOMBRE_CLIENTE con valor PUBLICO EN GENERAL

filtro1= df[(df["NOMBRE_CLIENTE"] == "PÚBLICO EN GENERAL")]
filtro1
```

```python
#Convertir el archivo filtrado a CSV
filtro1.to_csv("filtrado1(por objeto).csv")
```

```python
#Filtro por filas
#Se filtro filas del rango de 3 a 7, y se seleccionó unicamente la fila 2 y 5
filtro= df.iloc[3:7, :]
filtro2= df.iloc[[2,5], :]
filtro2
```

```python

#Convertir el archivo filtrado a CSV
filtro2.to_csv("filtrado2(por filas).csv")

```

```python
#Filtro por columnas
#se selecionó las columnas FECHA_DOC Y NOMBRE_VENDEDOR con todas sus filas
filtro3=df.iloc[:, 1:3]
filtro4=df.iloc[:, [1,4,5]]
filtro3
```

```python
#Convertir el archivo filtrado a CSV
filtro3.to_csv("filtro3(por columnas).csv")
```

```python
#Se filtro la columna 2 el cual empieza como columna 1
df1= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col=2)
df1
```

```python
#Convertir el archivo filtrado a CSV
df1.to_csv("Cambio de indice de columna.csv")
```

```python
#Paso 2: Aplicamos Filtro de filas vs columnas 
#Se selecciona las filas unicamente de AARON NUÑO con la columna CANTIDAD
filtro4= df1.loc[["AARON NUÑO CORDOVA"],["CANT"]]
filtro4
```

```python
#Convertir el archivo filtrado a CSV
filtro4.to_csv("filtro4 (filas VS columnas).csv")
```

```python
#Filtro por cabecera
#filtra las primeras 5 filas del DataFrame
filtro5 = df.head()
filtro5
```

```python
#Convertir el archivo filtrado a CSV
filtro5.to_csv("filtro5(por cabecera).csv")
```

```python
#Filtro por cola
filtro6= df.tail()
filtro6
```

**FILTROS BASICOS PARA VARIABLES CUANTITATIVAS**

```python
#Convertir el archivo filtrado a CSV
filtro6.to_csv("filtro6(por cola).csv")
```

```python
#Filtro por comparacion
#Selecciona solo las filas donde el valor en la columna "MARGEN_TOTAL_CALCULADO" es mayor a 10,000.
filtro7 = df[df["MARGEN_TOTAL_CALCULADO"] > 10000 ]
filtro7
```

```python
#Convertir el archivo filtrado a CSV
filtro7.to_csv("filtro7(por comparacion).csv")
```

```python
#Filtro Y
# Evalúa la columna "NOMBRE_CLIENTE",retorna True para las filas donde el cliente sea exactamente "MOLINERA DE MEXICO".
# y también evalúa la columna "FECHA_DOC",Retorna True para las filas donde la fecha sea exactamente "2022-01-05".
filtro8= df[(df["NOMBRE_CLIENTE"] ==  "MOLINERA DE MEXICO" )& (df["FECHA_DOC"] == "2022-01-05")]
filtro8
```

```python
#Convertir el archivo filtrado a CSV
filtro8.to_csv("filtro8(filtro &).csv")
```

```python
#filtro o
#Se filtra las filas que en la columna NOMBRE_CLIENTE sea MOLINERA DE MEXICO o ELICAMEx
filtro9= df[(df["NOMBRE_CLIENTE"] == "MOLINERA DE MEXICO") | (df["NOMBRE_CLIENTE"] == "ELICAMEX")]
filtro9
```

```python
#Convertir el archivo filtrado a CSV
filtro9.to_csv("filtro9(filtro o).csv")
```

```python
#Filtro not
#Se filtra tabla donde la columna NOMBRE_CLIENTE se descarta los datos que el cliente sea PUBLICO EN GENERAL
filtro10=df[~(df["NOMBRE_CLIENTE"] == "PÚBLICO EN GENERAL")]
filtro10
```

```python
#Convertir el archivo filtrado a CSV
filtro10.to_csv("filtro10(filtro not).csv")
```
